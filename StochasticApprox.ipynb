## Stochastic Approximation

Use stochastic iteration method to solve the root finding problem
$$g(w)=0$$
For two realistic reasons, you can not use regular gradient decent:
- only $\hat{g(w)}$ is available, which is $\hat{g}(w) = g(w) + \eta$
    
### RM algorithm
$$w_{k+1} = w_k - \alpha_k \hat{g}(w)$$



### TD-Learning


we have TD-learning algorithm






## sarsa TD for action value


```python
import numpy as np
import matplotlib.pyplot as plt
import random

def g_hat(w):
    return np.tanh(w-1) + random.gauss(3, 0.1)

a = 0.1
wk = 15 #initial value
wk_queue = []
g_hat_queue = []
for _ in range(1000):
    wk1 = wk - a * g_hat(wk)
    wk_queue.append(wk1)
    g_hat_queue.append(g_hat(wk1))
    wk = wk1
    if g_hat(wk1) < 1e-9:
        print("done!")
        break
        
plt.figure()
plt.plot(g_hat_queue)
plt.plot(wk_queue)
```
