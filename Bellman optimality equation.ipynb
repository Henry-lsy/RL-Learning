





### Example



```python
import numpy as np
import matplotlib.pyplot as plt

x_k=100 # initial condition
x_queue = []
for _ in range(100):
    x_k1 = 0.5 * np.sin(x_k)
    print(f"xk= {x_k}, fk={x_k1}")
    x_queue.append(x_k1)
    if abs(x_k1 - x_k)<1e-6:
        print("done!!")
        break
    x_k = x_k1
    
plt.figure()
plt.plot(x_queue)
```
